```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# style
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,6)

# reads
df = pd.read_csv('../data/processed/sales_clean.csv')

# kpi
total_revenue = df['total_price'].sum()
total_sales = df.shape[0]
avg_order = df['total_price'].mean()

print(f"Total Revenue: ${total_revenue:,.2f}")
print(f"Total Orders: {total_sales}")
print(f"Average Order Value: ${avg_order:,.2f}")

# sales/month
df['date'] = pd.to_datetime(df['date'])
monthly_sales = df.groupby(df['date'].dt.to_period('M'))['total_price'].sum()
monthly_sales.plot(kind='bar', color='skyblue', title='Monthly Revenue')
plt.show()

# Top 10
top_products = df.groupby('product_name')['total_price'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='barh', color='orange', title='Top 10 Products Revenue')
plt.show()

# Top sales by country
top_countries = country_sales.sort_values(ascending=False).head(10)
top_countries.plot(kind='barh', color='green', title='Top 10 Countries by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Country')
plt.tight_layout()
plt.show()

```
